Convert the columns "Created Date" and "Closed Date" to datetime datatype and create a new column "Request_Closing_Time" as the time elapsed between request creation and request closing.	

```python
import pandas as pd
#Checking the datatype of these columns before converting it to datetime.
print(df_CS_request['Created Date'].dtype)
print(df_CS_request['Closed Date'].dtype)
```

```python
# Converting date columns to datetime datatype
df_CS_request['Created Date'] = pd.to_datetime(df_CS_request['Created Date'], format='mixed')
df_CS_request['Closed Date'] = pd.to_datetime(df_CS_request['Closed Date'], format='mixed')
print(df_CS_request['Created Date'].dtype)
print(df_CS_request['Closed Date'].dtype)
```

```python
#Creating a new column as the duration between request creation and closing
df_CS_request['Request_Closing_Time']= df_CS_request['Closed Date'] - df_CS_request['Created Date']
df_CS_request['Request_Closing_Time']=df_CS_request['Request_Closing_Time'].dt.total_seconds() / 3600
df_CS_request['Request_Closing_Time']
```

```python
# Storing the date related columns in new dataframe.
df_date = df_CS_request[['Created Date', 'Closed Date', 'Request_Closing_Time']]
df_date
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Counting the frequency of each zip.
top_zips = df_CS_request['Incident Zip'].value_counts().head(10).index

# Filtering dataset for these ZIP codes
filtered_df = df_CS_request[df_CS_request['Incident Zip'].isin(top_zips)]

#Creating a pivot table of avg resolution time by ZIP and Complaint Type
pivot_table = filtered_df.pivot_table(
    index='Incident Zip',
    columns='Complaint Type',
    values='Request_Closing_Time',
    aggfunc='mean'
)

top_complaints = df_CS_request['Complaint Type'].value_counts().head(10).index
pivot_table = pivot_table[top_complaints.intersection(pivot_table.columns)]

#Plotting heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlOrRd', linewidths=0.5)
plt.title("Average Resolution Time by ZIP Code and Complaint Type")
plt.xlabel("Complaint Type")
plt.ylabel("Incident Zip")
plt.tight_layout()
plt.show()
```

●	Arrange the complaint types according to their average 'Request_Closing_Time', categorized by various locations. Illustrate it through graph as well.         

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping the data by complaint type and borough
grouped_data = df_CS_request.groupby(['Complaint Type', 'Borough'])['Request_Closing_Time'].mean().reset_index() # calculating mean request closing time

plt.figure(figsize=(10, 10))
#Creating a grouped barchart
sns.barplot(
    data=grouped_data,
    x='Request_Closing_Time',
    y='Complaint Type',
    hue='Borough',
    palette='Set2',
    )
#Adds the title to the barchart on the top
plt.title('Average Request Closing Time by Complaint Type and Borough')
plt.xlabel('Average Closing Time (hrs)')
plt.ylabel('Complaint Type')
plt.tight_layout()

#Displaying the created grouped barchart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Complaint Type and Location,and calculating average Request_Closing_Time
data = df_CS_request.groupby(['Complaint Type', 'Location Type'])['Request_Closing_Time'].mean().reset_index()
#Sorting the data in descending order
data = data.sort_values(by='Request_Closing_Time', ascending=False)

plt.figure(figsize=(12, 8))

#Converts the data into table format for plotting the heatmap
pivot_table = data.pivot(index='Complaint Type', columns='Location Type', values='Request_Closing_Time')
#Creating the heatmap using seaborn
sns.heatmap(pivot_table, cmap='YlOrRd', annot=True, fmt=".1f", linewidths=0.5) #cmap gives the color palette for heatmap
plt.title('Average Request Closing Time by Complaint Type and Location')
plt.xlabel('Location')
plt.ylabel('Complaint Type')
plt.tight_layout()

#Displaying the heatmap
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

grouped_times = [df_CS_request[df_CS_request['Complaint Type'] == c]['Request_Closing_Time']
                 for c in df_CS_request['Complaint Type'].unique()]
#Checking the null hypothesis
anova_result = stats.f_oneway(*grouped_times) #performing one-way ANOVA test

# Displaying the f-statistic result
print(f"F-statistic: {anova_result.statistic}")

#Checking the p vale using if-else condition
if anova_result.pvalue < 0.05:
    print("Reject the null hypothesis: There is major difference in response to mean time.")
else:
    print("Unable to reject the null hypothesis: There is no major difference in response to mean time.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

#creating contingency table
contingency_table = pd.crosstab(df_CS_request['Complaint Type'], df_CS_request [ 'Borough'])
#Performing chi-square test to see dependency between complaint type and borough
chi2=chi2_contingency(contingency_table)

#Displaying chi-square statistical value
print('Chi-square Statistic:', chi2[0])
#Checking the p value from chi-sqaure using if-else condition
if chi2[1]<0.05:
    print("Reject the null hypotheis: The complaint type and location are related to each other.")
else:
    print("Unable to reject the null hypotheis: The complaint type and location are not related to each other.")
```
